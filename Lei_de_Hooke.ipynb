```python
#Incertezas da Mola

delta_x1 = np.array([0.0236,0.28,0.276,0.295,0.305,0.30])
delta_x2= np.array([0.231,0.276,0.28,0.29,-0.301,0.305])
delta_x3= np.array([0.23,0.274,0.275,0.289,0.295,0.302])
```

```python
#plotando gráfico
# x representação a deformação da mola
x_1 = deformacaodamola_1 #dados coletados por emanuel
x_2=deformacaodamola_2 #dados coletados por Márcio
x_3= deformacaodamola_3#Dados coletados por Angelo

# y representa força newtoniana
y_1 = ForcaNewton_1
y_2=ForcaNewton_2
y_3= ForcaNewton_3
#dy representa a incerteza.
dy_1 = delta_x1  
dy_2=delta_x2
dy_3=delta_x3
```

```python
b,m=np.polynomial.polynomial.polyfit(x_1,y_1,1,w=dy)
b_2,m_2=np.polynomial.polynomial.polyfit(x_2,y_2,1,w=dy_2)
b_3,m_3=np.polynomial.polynomial.polyfit(x_3,y_3,1,w=dy_3)
#w=dy
```

```python
funcaolinear= m*x_1+b
funcaolinear_2= m_2*x_2+b_2
funcaolinear_3= m_3*x_3+b_3
```

```python

def Delta(x_1, dy):
    D = (sum(1/dy**2))*(sum(x_1**2/dy**2))-(sum(x_1/dy**2))**2
    return D

D=Delta(x_1, dy)

dm = math.sqrt(1/D*sum(1/dy**2)) #error in slope
db = math.sqrt(1/D*sum(x_1**2/dy**2)) #error in intercept






# Márcio Henrique
def Delta_2(x_2, dy_2):
    D = (sum(1/dy_2**2))*(sum(x_1**2/dy_2**2))-(sum(x_2/dy_2**2))**2
    return D

D_2=Delta(x_2, dy_2)

dm_2 = math.sqrt(1/D_2*sum(1/dy_2**2)) #error in slope
db_2 = math.sqrt(1/D_2*sum(x_2**2/dy_2**2)) #error in intercept


#angelo

def Delta_3(x_3, dy_3):
    D = (sum(1/dy_3**2))*(sum(x_1**2/dy_3**2))-(sum(x_3/dy_3**2))**2
    return D

D_3=Delta(x_3, dy_3)

dm_3 = math.sqrt(1/D_3*sum(1/dy_3**2)) #error in slope
db_3 = math.sqrt(1/D_3*sum(x_3**2/dy_3**2)) #error in intercept
```

```python
def LLSFD2(x_1,y_1,dy):
    N = sum(((y_1-b-m*x_1)/dy)**2)
    return N
                      
N = LLSFD2(x_1,y_1,dy)





#Márcio



def LLSFD2_2(x_2,y_2,dy_2):
    N_2 = sum(((y_2-b_2-m*x_2)/dy_2)**2)
    return N_2
                      
N = LLSFD2(x_2,y_2,dy_2)


#angelo

def LLSFD2_3(x_3,y_3,dy_3):
    N_2 = sum(((y_3-b_3-m*x_3)/dy_3)**2)
    return N_3
                      
N = LLSFD2(x_3,y_3,dy_3)
```
