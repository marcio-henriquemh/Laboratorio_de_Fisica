**Descrição do Movimento **


𝑥(𝑡) = |𝑣0|.cos(𝜃).𝑡 <br />


𝑦(𝑡) = |𝑣0|.sen(𝜃).𝑡 − 𝑔𝑡² 2


O ângulo de lançamento (𝜃) deve ser lido do usuário (garanta que seja um valor válido). <br />
• A velocidade inicial de lançamento (𝑣0) deve ser lida do usuário (garanta que seja um valor válido). <br />
• Deve-se variar o tempo (𝑡) em incrementos de 0,1 s. <br />
• Deve-se assumir 𝑔 = 9,8 𝑚/𝑠² (aceleração da gravidade no planeta Terra). <br />
• Além de apresentar na tela o gráfico da trajetória do projétil, deve-se apresentar também: 
a distância total percorrida pelo mesmo (em metros), a altura máxima atingida (em metros) e a duração total do lançamento (em segundos)

```python
#bibliotecas

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Dados coletados das Alturas e Lancamentos em graus.

#gravidade
g=9.8
#Lendo os dados

#dados altura 1
tempo=np.array([0.3587,0.3540,0.3541])
altura=np.array([0.212,0.212,0.205])
anguloradiano=np.deg2rad(np.array([altura[0],altura[1],altura[2]]))


alcanceMax = (np.array([tempo[0]**2,tempo[1]**2,tempo[2]**2])) * (np.sin(2*np.array([anguloradiano[0],anguloradiano[1],anguloradiano[2]])) / g)
alturaMax = (np.array([tempo[0]**2,tempo[1]**2,tempo[2]**2])) * np.sin(np.array([anguloradiano[0],anguloradiano[1],anguloradiano[2]]))**2 / (2*g)
tempocompleto = (2*np.array([tempo[0],tempo[1],tempo[2]]))* np.sin(np.array([anguloradiano[0],anguloradiano[1],anguloradiano[2]]) / g)

print(alcanceMax)
print(alturaMax)
print(tempocompleto)
#representar em vetores os dados em lançamento
```

```python
x = abs(np.array([tempo[0],tempo[1],tempo[2]])) * np.cos(np.array([anguloradiano[0],anguloradiano[1],anguloradiano[2]])) * tempocompleto
y=abs(np.array([tempo[0],tempo[1],tempo[2]])) * np.sin(np.array([anguloradiano[0],anguloradiano[1],anguloradiano[2]])) * tempocompleto - ((g*(tempocompleto**2))/2)
print(x)

print(y)
```

```python
plt.title("Trajetória do Projétil")
plt.xlabel("Distância (m)")
plt.ylabel("Altura (m)")
#plt.plot(np.array([x[0],x[1],x[2]]))#,(np.array([y[0],y[1],y[2]]))#,'r')
plt.plot(np.array([x[0],x[1],x[2]]))
plt.show()



```

```python
plt.title("Trajetória do Projétil")
plt.xlabel("Distância (m)")
plt.ylabel("Altura (m)")
plt.plot(np.array([y[0],y[1],y[2]]))
plt.show()
plt.show()
```
